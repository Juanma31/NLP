```python
#Función que dado un dataframe y una columna, devuelve un diccionario con las palabras, bigramas y trigramas mas usados en la columna
def get_vocabulary(df, col_name):
    
    words_ = []
    bigrams_ = []
    trigrams_ = []
    
    for review in df[col_name]:
        for word in review.split():
            words_.append(word)
    
    bigrams_ = list(ngrams(words_, 2))
    trigrams_ = list(ngrams(words_, 3))
    
    vocab = {
        'words': words_,
        'bigrams': bigrams_,
        'trigrams': trigrams_
    }
    
    return vocab
    
```

```python
automotive_vocab = get_vocabulary(df_automotive, 'reviewText')
video_vocab = get_vocabulary(df_video, 'reviewText')
instruments_vocab = get_vocabulary(df_instruments, 'reviewText')
garden_vocab = get_vocabulary(df_garden, 'reviewText')
```

```python
#Preprocesamiento del texto
def text_normalizer(text, lemmas_dict=None, sw_list=None):

    processed_text = []
    
    if sw_list is None:
        sw_list = []
    
    # Convertimos el texto a minúsuculas
    text = text.lower()
    
    # Eliminamos números
    text = ''.join([c for c in text if not c.isdigit()])
    
    # Eliminamos caracteres "extraños"
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    # Creamos tabla para eliminar signos de puntuación
    table = str.maketrans('', '', string.punctuation)
    
    # Segmentamos texto en frases
    sentences = text.split('.')
    
    # Para cada frase
    for sentence in sentences:
        words = sentence.split(' ')
        # Para cada palabra
        for word in words:
            if word not in sw_list:
                if word not in string.punctuation:
                    word = word.translate(table)
                    if lemmas_dict is not None:
                        word = lemmas_dict.get(word, word)
                processed_text.append(word)
    
    return ' '.join(processed_text)
```

```python
# Funcion para guardar, en una ruta dada, un diccionario de palabras lematizadas
def get_lemmas_dict(data_path, lemmas_dict_file):
    lemmas_dict = {}
    with open(os.path.join(data_path, lemmas_dict_file), 'r') as f:
        for line in f:
            (key, val) = line.split()
            lemmas_dict[str(val)] = key
    return lemmas_dict
```

```python
lemmas_path = '../data'
english_lemmas = 'lemmatization-en.txt'  # Fuente: https://github.com/michmech/lemmatization-lists

english_lemmas_dict = get_lemmas_dict(lemmas_path, english_lemmas)
```

```python
sw_list = get_stop_words('english')
```

```python
# Creamos del texto procesado
df_automotive['processed_text'] = df_automotive['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_video['processed_text'] = df_video['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_instruments['processed_text'] = df_instruments['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_garden['processed_text'] = df_garden['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
```

# Modelado de topics

```python
#Entrenamos un LDA para el modelado de topics
#Diccionario id-término y matriz documento-palabra
all_reviews_text = \
    list(df_automotive['reviewText'].values) + \
    list(df_video['reviewText'].values) + \
    list(df_instruments['reviewText'].values) + \
    list(df_garden['reviewText'].values)

all_reviews_text = [review.split() for review in all_reviews_text]
```

```python
all_reviews_processed_text = \
    list(df_automotive['processed_text'].values) + \
    list(df_video['processed_text'].values) + \
    list(df_instruments['processed_text'].values) + \
    list(df_garden['processed_text'].values)

all_reviews_processed_text = [review.split() for review in all_reviews_processed_text]
```

```python
raw_dictionary = Dictionary(all_reviews_text)
processed_dictionary = Dictionary(all_reviews_processed_text)
```

```python
raw_corpus = [raw_dictionary.doc2bow(doc) for doc in all_reviews_text]
processed_corpus = [processed_dictionary.doc2bow(doc) for doc in all_reviews_processed_text]
```

```python
#Elegimos un número de topics igual a cuatro 
#ya que es número de categorías que estamos usando como corpus. 
num_topics = 4
```

```python
lda_model_raw = LdaModel(
    corpus=raw_corpus,
    id2word=raw_dictionary,
    num_topics=num_topics,
    iterations=5,
    passes=10,
    alpha='auto'
)
```

```python
word_dict_raw = {}

for i in range(num_topics):
    words = lda_model_raw.show_topic(i, topn = 20)
    word_dict_raw['Topic #' + '{:02d}'.format(i+1)] = [i[0] for i in words]
pd.DataFrame(word_dict_raw)
```

A priori no se identifica ningún topic

```python
lda_model_processed = LdaModel(
    corpus=processed_corpus,
    id2word=processed_dictionary,
    num_topics=num_topics,
    iterations=5,
    passes=10,
    alpha='auto'
)
```

```python
word_dict_processed = {}

for i in range(num_topics):
    words = lda_model_processed.show_topic(i, topn = 15)
    word_dict_processed['Topic #' + '{:02d}'.format(i+1)] = [i[0] for i in words]
pd.DataFrame(word_dict_processed)
```

- El topic 1 hace referencias a cosas relacionada con los automoviles (feeder, battery)
- El topic 2 parece que hacer referecia a cosas de patio, cesped y jardin (grill)
- El topic 3 parece que hace referencia a cosas relacionadas con Amazon Instant Video (show, season, episode, series, character)
- El topic 4 parece que hace referencia a cosas relacionadas con la musica (guitar, strings, sound)

# Gráficos descriptivos

```python
# Perplexity
print('\nPerplexity (without text processing): ', lda_model_raw.log_perplexity(raw_corpus))

# Coherence Score
coherence_model_lda_raw = CoherenceModel(model=lda_model_raw, texts=all_reviews_text, dictionary=raw_dictionary, coherence='c_v')
coherence_lda_raw = coherence_model_lda_raw.get_coherence()
print('\nCoherence Score (without text processing): ', coherence_lda_raw)
```

```python
# Perplexity
print('\nPerplexity (with text processing): ', lda_model_processed.log_perplexity(processed_corpus))

# Coherence Score
coherence_model_lda_processed = CoherenceModel(model=lda_model_processed, texts=all_reviews_processed_text, dictionary=processed_dictionary, coherence='c_v')
coherence_lda_processed = coherence_model_lda_processed.get_coherence()
print('\nCoherence Score (with text processing): ', coherence_lda_processed)
```

```python
# Buscamos un número de topics - teóricamente - mejor
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        
        # Build LDA model
        model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=num_topics)
        
        # Create a list of LDA models
        model_list.append(model)
        
        # Compute the Coherence for each model
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
start_ = 1
end_ = 20
step_ = 1
```

```python
model_list, coherence_values = compute_coherence_values(
    dictionary=processed_dictionary,
    corpus=processed_corpus,
    texts=all_reviews_processed_text,
    start=start_,
    limit=end_,
    step=step_
)
```

```python
x = range(start_, end_, step_)
plt.plot(x, coherence_values)
plt.xlabel('Num Topics')
plt.ylabel('Coherence score')
plt.legend(('coherence_values'), loc='best')
plt.show()
```
